```python
# Выделяем целые удары сердца
for i in range (len(df_number_csv)):


    number = df_number_csv.iloc[i]['number']

    path_data = f"data\MLII\{number}MLII.csv"
    df_data = pd.read_csv(path_data, sep=",")

    last_heart_beat = df_data.loc[len(df_data)-1,'heart_beat']
    df_data = df_data[(df_data.heart_beat != 0)&(df_data.heart_beat != 1)&(df_data.heart_beat != last_heart_beat)]
    df_data.heart_beat -= 1
    

    path_save = f"data\MLII(whole_beats)\{number}MLII.csv"
    df_data.to_csv(path_save, sep=',', index=False)
```

```python
# Склеиваем по 15 ударов с каждой ЭКГ удаляя неправильные удары 

for i in range (len(df_number_csv)):


    number = df_number_csv.iloc[i]['number']

    path_data = f"data\MLII(whole_beats)\{number}MLII.csv"
    df_data = pd.read_csv(path_data, sep=",")
    if i == 0:
        df_data_all = df_data[(df_data.heart_beat < 16)]
    else:
        df_data = df_data[(df_data.heart_beat < 16)]
        df_data_all = pd.concat([df_data_all, df_data], ignore_index=True)
    
df_data_all = df_data_all[(df_data_all.Type == 'N')]

display(df_data_all)

path_save = f"data\DataTrain\TrainMLII(NS).csv"
df_data_all.to_csv(path_save, sep=',', index=False)
```
