```python
# Работаем с первым столбцом
df_number_csv = pd.DataFrame({'number': [104, 105, 106, 107, 108, 109, 111, 112, 113, 114 , 115, 116, 117, 118, 119, 121, 122, 123 , 124, 200, 201, 202, 203, 205, 207, 208, 209, 210, 212, 213, 214, 215,  217, 219, 220, 221, 222, 223, 228, 230, 231, 232, 233, 234]})
df_number_csv

for i in range (len(df_number_csv)):


    number = df_number_csv.iloc[i]['number']

    path_annotations = f"data\{number}annotations.txt"
    df_annotations = pd.read_table(path_annotations, sep='\s+', engine='python')

    path_data = f"data\{number}.csv"
    df_data = pd.read_csv(path_data, sep=",")

    if df_data.columns[1] == "'MLII'":

        # удаляем столбец
        df_data.drop (columns=df_data.columns [2], axis= 1 , inplace= True )

        df_data['Type'] = np.nan
        df_data['heart_beat'] = np.nan

        last_sample = 0
        
        for i in range(len(df_annotations)):

            if i < 2:
                for g in range(last_sample, df_annotations.iloc[i]['Sample#']):
                    df_data.loc[g,'Type'] = df_annotations.iloc[i]['Type']
                    df_data.loc[g,'heart_beat'] = i
            
            else:

                j = i -1
                for g in range(last_sample, df_annotations.iloc[i]['Sample#']):
                    df_data.loc[g,'Type'] = df_annotations.iloc[j]['Type']
                    df_data.loc[g,'heart_beat'] = i

            last_sample = df_annotations.iloc[i]['Sample#']
            last_type = df_annotations.iloc[i]['Type']

        for i in range(last_sample, len(df_data)):
            df_data.loc[i,'Type'] = last_type
            df_data.loc[i,'heart_beat'] = df_data.loc[last_sample-1,'heart_beat']+1

        path_save = f"data\MLII\{number}MLII.csv"
        df_data.to_csv(path_save, sep=',', index=False)

    if df_data.columns[1] == "'V5'":

        # удаляем столбец
        df_data.drop (columns=df_data.columns [2], axis= 1 , inplace= True )

        df_data['Type'] = np.nan
        df_data['heart_beat'] = np.nan

        last_sample = 0
        
        for i in range(len(df_annotations)):

            if i < 2:
                for g in range(last_sample, df_annotations.iloc[i]['Sample#']):
                    df_data.loc[g,'Type'] = df_annotations.iloc[i]['Type']
                    df_data.loc[g,'heart_beat'] = i
            
            else:

                j = i -1
                for g in range(last_sample, df_annotations.iloc[i]['Sample#']):
                    df_data.loc[g,'Type'] = df_annotations.iloc[j]['Type']
                    df_data.loc[g,'heart_beat'] = i

            last_sample = df_annotations.iloc[i]['Sample#']
            last_type = df_annotations.iloc[i]['Type']

        for i in range(last_sample, len(df_data)):
            df_data.loc[i,'Type'] = last_type
            df_data.loc[i,'heart_beat'] = df_data.loc[last_sample-1,'heart_beat']+1

        path_save = f"data\V5\{number}V5.csv"
        df_data.to_csv(path_save, sep=',', index=False)

        
```
